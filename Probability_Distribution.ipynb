Ans Q5--

Kurtosis

Kurtosis is a statistical measure that describes the shape of a distribution, especially:

the peakedness
the flatness
and the presence of extreme values (tails/outliers)

It helps us understand how data is distributed compared to a normal distribution.

Mesokurtic (normal)


Leptokurtic (sharp peak, heavy tails)


Platykurtic (flat peak, light tails)




Ans Q6--

A uniform distribution is a probability distribution in which all outcomes have an equal chance of occurring.

Rolling a fair die is a good example of a uniform distribution because:

A fair die has 6 possible outcomes:

1,2,3,4,5,6
Each number has the same probability of occurring.
